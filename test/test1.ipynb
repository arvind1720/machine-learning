```python
columns = ['Age', 'Gender', 'Polyuria', 'Polydipsia', 'Sudden Weight Loss', 'Weakness', 'Polyphagia', 'Genital thrush',
           'visual blurring', 'Itching', 'Irritability', 'delayed healing', 'partial paresis', 'muscle stiffness',
           'Alopecia', 'Obesity', 'Class_labels']

# Assuming 'df' is your DataFrame read from the CSV file
df = pd.read_csv('diabetes_risk_prediction_dataset.data.txt', names=columns)

# Mapping dictionary for binary columns
mapping = {"Yes": 1, "No": 0}

# Mapping dictionary for 'Gender'
gender_mapping = {"Male": 1, "Female": 0}

# Columns to apply the mapping
columns_to_map = ['Polyuria', 'Polydipsia', 'Sudden Weight Loss', 'Weakness', 'Polyphagia', 'Genital thrush',
                  'visual blurring', 'Itching', 'Irritability', 'delayed healing', 'partial paresis',
                  'muscle stiffness', 'Alopecia', 'Obesity']

# Applying mapping to specified columns
df[columns_to_map] = df[columns_to_map].apply(lambda col: col.map(mapping))

# Mapping 'Gender' column
df['Gender'] = df['Gender'].map(gender_mapping)

# Displaying the updated DataFrame
df.head()
```

```python
df.describe()
```

```python
#sns.pairplot(df, hue='Class_labels')
```

```python
data=df.values

X=data[:,0:16]
Y=data[:,16]
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

VALIDATION DATA

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15,random_state=42)
```

PREDICTION ALGORITHM 1: DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier  # Importing the DecisionTreeClassifier from the scikit-learn library for building a decision tree model.

model_DTC = DecisionTreeClassifier()  # Creating an instance of the DecisionTreeClassifier as model_DTC.

model_DTC.fit(X_train, y_train)  # Training the model on the training data X_train and corresponding labels y_train.
```

```python
prediction_test1=model_DTC.predict(X_test)             # Predicting labels for the test data using the trained DecisionTreeClassifier model.


#Calculating Testing data accuracy
from sklearn.metrics import accuracy_score
print("   \nDECISION TREE Performance  Metrics")
print("\nAccuarcy of testing data = ",accuracy_score(y_test,prediction_test1))


#Calculate Training data accuracy
prediction_train1=model_DTC.predict(X_train)
accuracy_train1 = accuracy_score(y_train, prediction_train1)
print("\nAccuracy of training data =", accuracy_train1)



#Calculate Validation data accuracy
prediction_val1=model_DTC.predict(X_val)
accuracy_val1 = accuracy_score(y_val, prediction_val1)
print("\nAccuracy on validation data =", accuracy_val1)





```

PREDICTION ALGORITHM 2: KNN

```python
from sklearn.neighbors import KNeighborsClassifier  # Importing the KNeighborsClassifier from scikit-learn for K-Nearest Neighbors algorithm.

model_knn = KNeighborsClassifier(5)  # Creating an instance of KNeighborsClassifier with 5 neighbors.

model_knn.fit(X_train, y_train)  # Training the KNN model on the training data X_train and corresponding labels y_train.
```

```python
prediction_test2=model_knn.predict(X_test)      # Predicting labels for the test data using the trained KNN model.



#Calculate testing dataaccuracy
from sklearn.metrics import accuracy_score
print("   \nKNN Performance  Metrics")
print("\nAccuarcy of testing data  = ",accuracy_score(y_test,prediction_test2))

#Calculate Training data accuracy
prediction_train2=model_knn.predict(X_train)
accuracy_train2 = accuracy_score(y_train, prediction_train2)
print("\nAccuracy of training data =", accuracy_train2)


#Calculate Validation data accuracy
prediction_val2=model_knn.predict(X_val)
accuracy_val2 = accuracy_score(y_val, prediction_val2)
print("\nAccuracy on validation data =", accuracy_val2)




```

CLASSIFICATION REPORT OF TESTING DATA

```python
from sklearn.metrics import classification_report,precision_recall_curve
print("                    DECISON TREE ALGORITHM            ")
print(classification_report(y_test,prediction_test1))



print("                    KNN ALGORITHM            ")
print(classification_report(y_test,prediction_test2))
```
